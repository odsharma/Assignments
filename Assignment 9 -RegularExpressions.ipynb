#### Q1: Write regular expression to match email address

```python
import re
with open("mbox-short.txt") as fh:
    content = fh.read()
    email=re.findall("[a-zA-Z0-9]*@[a-zA-Z]+\.[a-zA-Z\.]+",content)
    print(email)
```

#### Q2: Write regular expression to match URL

```python
import re
with open("mbox-short.txt") as fh1:
    content = fh1.read()
    url=re.findall("http://[\S]+|https://[\S]+",content)
    print(url)
```

#### Q3: Build a regular expression to extract URLs from the server logs 'access.log.41'

```python
import re
with open("access.log.41") as fh2:
    content=fh2.read()
    url=re.findall("GET /\S+|POST /\S+",content)
    print(url)
```

#### Q4.Given an input string (s) and a pattern (p), implement wildcard pattern matching with support for '?' and '*' where:

'?' Matches any single character.
'*' Matches any sequence of characters (including the empty sequence).
The matching should cover the entire input string (not partial).

```python
    def isMatch(s, p) -> bool:
        if len(p) ==0 and  len(s)==0:
            return True
        if len(p)>0:
            
            if p[0] == "?":
                if len(s) ==0:
                    return False
                return isMatch(s[1:], p[1:])
            elif p[0] == "*":
                if len(p)>1:
                    for i in range(len(s)):
                        if p[1] == s[i]:
                            return isMatch(s[i:], p[1:])
                    return False
                else: 
                    return True
            elif p[0] == s[0]:
                return isMatch(s[1:], p[1:])
            elif p[0] != s[0]:
                return False
            if len(s)==0 and len(p)>0:
                return False
        elif len(p) == 0 or len(s) == 0:
            return False
```
